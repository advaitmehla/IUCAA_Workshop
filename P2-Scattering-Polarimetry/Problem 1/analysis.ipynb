```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df10 = pd.read_csv('data/out_10.csv',sep = ";")
df20 = pd.read_csv('data/out_20.csv',sep = ";")
df50 = pd.read_csv('data/out_50.csv',sep = ";")
df100 = pd.read_csv('data/out_100.csv',sep = ";")
```

```python
def az(dire):
    spl = dire.split(',')
    x = float(spl[0][1:])
    y = float(spl[1])
    return(np.arctan(y/x))

def azimuths(df):
    comp = df[df['Process'] == 'compt']['Scattered Direction']
    ral = df[df['Process'] == 'Rayl']['Scattered Direction']
    az_comp = [az(comp.iloc[i]) for i in range(len(comp))]
    az_ral = [az(ral.iloc[i]) for i in range(len(ral))]
    return az_comp, az_ral
```

```python
comp10, ral10 = azimuths(df10)
comp20, ral20 = azimuths(df20)
comp50, ral50 = azimuths(df50)
comp100, ral100 = azimuths(df100)
```

```python
plt.figure(figsize = (10,8))
plt.hist(comp10,histtype = 'step',bins =100 ,density =True,label = "Comp 10");
plt.hist(comp20,histtype = 'step',bins =100 ,density =True,label = "Comp 20");
plt.hist(comp50,histtype = 'step',bins =100 ,density =True,label = "Comp 50");
plt.hist(comp100,histtype = 'step',bins =100 ,density =True,label = "Comp 100");
plt.legend()
```

```python
plt.figure(figsize = (10,8))
plt.hist(ral10,histtype = 'step',bins =100 ,density =True,label = "ral 10");
plt.hist(ral20,histtype = 'step',bins =100 ,density =True,label = "ral 20");
plt.hist(ral50,histtype = 'step',bins =100 ,density =True,label = "ral 50");
plt.hist(ral100,histtype = 'step',bins =100 ,density =True,label = "ral 100");
plt.legend()
```
